```python
df['Sale_LastMonth']=df['Sales'].shift(+1)
df['Sale_2Monthsback']=df['Sales'].shift(+2)
df['Sale_3Monthsback']=df['Sales'].shift(+3)
df
```

```python
from sklearn.linear_model import LinearRegression
lin_model=LinearRegression()
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,max_features=3, random_state=1)
```

```python
x1,x2,x3,y=df['Sale_LastMonth'],df['Sale_2Monthsback'],df['Sale_3Monthsback'],df['Sales']
x1,x2,x3,y=np.array(x1),np.array(x2),np.array(x3),np.array(y)
x1,x2,x3,y=x1.reshape(-1,1),x2.reshape(-1,1),x3.reshape(-1,1),y.reshape(-1,1)
final_x=np.concatenate((x1,x2,x3),axis=1)
print(final_x)
```

```python
X_train,X_test,y_train,y_test=final_x[:-30],final_x[-30:],y[:-30],y[-30:]
```

```python
model.fit(X_train,y_train)
lin_model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,8)#Adjust figure parameter
plt.plot(pred,label='Random_Forest_Predictions')# Show Prediction values 
plt.plot(y_test,label='Actual Sales')# Show Y test valus
plt.legend(loc="upper left")
plt.title("Forecasting Sales & Actual Sales using Random Forest")

plt.show()
```

```python
lin_pred=lin_model.predict(X_test)
plt.rcParams["figure.figsize"] = (11,6)
plt.plot(lin_pred,label='Linear_Regression_Predictions')
plt.plot(y_test,label='Actual Sales')
plt.legend(loc="lower right")
plt.title("Forecasting Sales & Actual Sales using Linear Regression")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
RMSE_RandomForest=sqrt(mean_squared_error(pred,y_test))
RMSE_linearRegression=sqrt(mean_squared_error(lin_pred,y_test))
```

```python
print('Mean Squared Error for Random Forest Model is:',RMSE_RandomForest)
print('Mean Squared Error for Linear Regression Model is:',RMSE_linearRegression)
```
